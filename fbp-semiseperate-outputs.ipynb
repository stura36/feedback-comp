## Vocabulary

```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def word_stem(words:list):
    '''
    words - list of words in a text
    
    returns - list of stemmed words
    '''
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in words]
    return stemmed_words

def filter_stop_words(tokens:list):
    '''
    tokens - list of tokens
    
    returns - list of words without stop words
    '''
    return [w for w in tokens if not w.lower() in stop_words]

def filter_puncts(tokens:list):
    '''
    tokens - list of tokens
    
    returns - list of tokens without tokens of punctuation marks
    '''
    return [w for w in tokens if not w in ".,:!?"]

def text_to_words(data):
    '''
    data - input data
    
    Removes stop words, punctuation marks and applys stemming on input text.
    
    returns - dataframe of preprocessed data
    '''
    
    data_tokenized = data["full_text"].apply(nltk.word_tokenize)
    data_filtered  = data_tokenized.apply(filter_stop_words)
    data_filtered = data_tokenized.apply(filter_puncts)
    data_stemmed = data_tokenized.apply(word_stem)
    
    return data_stemmed

def cnt_uniq(tokens:list):
    '''
    tokens - list of tokens
    
    returns - number of uniq tokens in a text
    '''
    return len(set(tokens))

def text_to_words_all(data):
    data_tokenized = data["full_text"].apply(nltk.word_tokenize)
    data_filtered = data_tokenized.apply(filter_puncts)
    return data_filtered

def mean_word_len(text_tokens):
    mean_len = np.mean([len(w) for w in text_tokens])
    return mean_len

def voc_pipe(data):
    prepared_text = text_to_words(data)
    uniq_cnts = prepared_text.apply(cnt_uniq)
    prepared_text_all = text_to_words_all(data)
    text_means = prepared_text_all.apply(mean_word_len)
    
    return [text_means,uniq_cnts]
```

```python
def text_to_sentences(text):
    data_tokens = nltk.word_tokenize(text)
    sentences = list()
    last = 0
    flag = 1
    for i,tkn in enumerate(data_tokens):
        if tkn in "!?.":
            sentence = " ".join(data_tokens[last:i])
            last = i+1
            sentences.append(sentence)
            flag = 0
    if flag == 1:
        sentence = " ".join(data_tokens)
        sentences.append(sentence)
        
    return sentences
```

```python
from sentence_transformers import util
from collections import deque

def calc_emb_dist(emb_text):
    num_of_dist = len(emb_text)-1
    distances = np.array(util.cos_sim(emb_text,emb_text))
    idxs1 = np.array([i for i in range(num_of_dist)])
    idxs2 = idxs1+1
    try:
        unpad_output = distances[idxs1,idxs2]
        padded_output = np.pad(unpad_output,(0,max_len-len(emb_text)))        
    except IndexError:
        padded_output = np.zeros(max_len-1)
    return padded_output

```

## MODEL


```python
BATCH_SIZE = 16
MAX_LEN = 256
DROPOUT = 0.1 # 0.2
LEARNING_RATE = 1e-5
EPOCHS = 3#8
AUTO = tf.data.experimental.AUTOTUNE
MODEL = "deberta"
```

```python
from transformers import DebertaTokenizer
tokenizer = DebertaTokenizer.from_pretrained("microsoft/deberta-base")
```

```python

def encode_texts(text):
    
    encoded = tokenizer.batch_encode_plus(text, add_special_tokens = True, max_length= MAX_LEN,
              padding='max_length', truncation=True, return_attention_mask=True)

    input_ids       = np.array(encoded["input_ids"], dtype="int32")
    attention_masks = np.array(encoded["attention_mask"], dtype="int32")

    return input_ids, attention_masks

        
```

```python
class CLSEmbeddingLayer(tf.keras.layers.Layer):
    def __init__(self):
        super(CLSEmbeddingLayer, self).__init__()
    
    def call(self, last_hidden_state):
        last_hidden_state = outputs[0]
        cls_embeddings = last_hidden_state[:, 0]
        return cls_embeddings

class MeanPoolingLayer(tf.keras.layers.Layer):
    def __init__(self):
        super(MeanPoolingLayer, self).__init__()
    
    def call(self, last_hidden_state, attention_mask):
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()
        sum_embeddings = torch.sum(last_hidden_state * input_mask_expanded, 1)
        sum_mask = input_mask_expanded.sum(1)
        sum_mask = torch.clamp(sum_mask, min=1e-9)
        mean_embeddings = sum_embeddings / sum_mask
        return mean_embeddings
    
class MaxPoolingLayer(tf.keras.layers.Layer):
    def __init__(self):
        super(MaxPoolingLayer, self).__init__()
    
    def call(self, last_hidden_state, attention_mask):
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(last_hidden_state.size()).float()
        last_hidden_state[input_mask_expanded == 0] = -1e9  # Set padding tokens to large negative value
        max_embeddings = torch.max(last_hidden_state, 1)[0]
        return max_embeddings
```

```python
from transformers import TFDebertaModel
from tensorflow.keras import layers
from tensorflow.keras.layers import Dropout, Dense

def mcrmse(y_true, y_pred):
    '''Returns value for MCRMSE (competition's metric)'''
    
    colwise_mse = tf.reduce_mean(tf.square(y_true - y_pred), axis=0)
    return tf.reduce_mean(tf.sqrt(colwise_mse), axis=0)

def build_model(): 
    '''Builds DeBerta based model.'''
    
    model_ids  = layers.Input(shape=(MAX_LEN, ), dtype = tf.int32, name = "input_ids")
    model_mask = layers.Input(shape=(MAX_LEN, ), dtype = tf.int32, name = "attention_mask")
    
    transformer_box = TFDebertaModel.from_pretrained("microsoft/deberta-base")
    
    sequence_output = transformer_box(input_ids = model_ids, attention_mask = model_mask)[0]       
    
    
    clf_output = sequence_output[:,0,:]
    clf_output = Dropout(DROPOUT)(clf_output)
    
    intermediate_layer = Dense(MAX_LEN,activation = "relu", name = "intermediate_layer")(clf_output)
    
    
    outputs = Dense(4, name = "output_layer")(intermediate_layer)
    
    
    
        
    model = tf.keras.Model(inputs = [model_ids, model_mask], outputs = outputs)
    
    # Optimizer
    optimizer = tf.keras.optimizers.Adam(learning_rate = LEARNING_RATE)

    # Compile the model
    model.compile(optimizer = optimizer , loss = mcrmse, metrics=[mcrmse])
    
    return model
```

```python
model = build_model()
```

```python
X_train_enc = encode_texts(X_train)
X_test_enc = encode_texts(X_test)


```

```python
train_dataset = (
            tf.data.Dataset
            .from_tensor_slices((X_train_enc, y_train))
            .repeat()
            .shuffle(X_train_enc[0].shape[0])
            .batch(BATCH_SIZE)
            .prefetch(AUTO)
        )
valid_dataset = (
            tf.data.Dataset
            .from_tensor_slices((X_test_enc, y_test))
            .batch(BATCH_SIZE)
            .cache()
            .prefetch(AUTO)
        )
```
